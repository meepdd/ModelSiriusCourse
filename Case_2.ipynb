Функции поиска звуковой характеристики, чтения и загрузки звуковых характеристик

```python
def get_sound_features(audio):
    # Compute spectrogram
    D = librosa.stft(audio)
    # Convert power spectrogram to dB scale
    S_db = librosa.amplitude_to_db(abs(D), ref=np.max)
    return S_db.mean(axis=1)
```

```python
def read_and_extract_features(filename):
    audio_data, sr = librosa.load(filename, sr=None)
    return get_sound_features(audio_data)
```

Выделение различных форматов, чтобы библиотека librosa не выдавала ошибку.
Удаляем постоянно загружающийся файл из-за особенности ПО - DS_Store

```python
sound_formats = ['mp3', 'aac', 'amr', 'm4a', 'wav']

def is_audio_file(fname):
    return fname[-3:] in sound_formats

def get_all_filenames(path):
    folders = os.listdir(path)
    folders.remove('.DS_Store')
    filenames = []
    for folder in folders:
        new_files = os.listdir(os.path.join(path, folder))
        filenames.extend([os.path.join(path, folder, filename) for filename in new_files])
    filenames = list(filter(is_audio_file, filenames))
    return filenames
```

Добавляем training samples

```python
train_path = '/Users/md/Downloads/DataSet/training_samples/'
```

```python
filenames = get_all_filenames(train_path)
filenames
```

```python
train_data = [read_and_extract_features(fname) for fname in tqdm(filenames)]
```

Превращаем датафайлы в array

```python
X_train_raw = np.array(train_data)
```

Библиотеки

```python
from sklearn.svm import LinearSVC
from xgboost import XGBClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.model_selection import train_test_split
```

```python
model = RandomForestClassifier()
```

```python
# Разделить train на train and validation

X_train, X_val, y_train, y_val = train_test_split(
    X_train_raw,
    labels,
    random_state=42,
    shuffle=True,
    test_size=0.25
)
```
